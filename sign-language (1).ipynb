```python
import os
import tensorflow as tf
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import numpy as np
import PIL
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras import layers
```

```python
def get_images(images, folder, arr, label):
    temp2 = os.listdir(folder)
    limit = 600
    count = 0
    for y in temp2:
        if count < limit:
            img = cv2.imread(os.path.join(folder, y))
            img = cv2.resize(img, (80, 80))
            if img is not None:
                images.append(img)
                arr.append(label)
                count += 1
        else:
            break
```

```python
images = []
labels = []
get_images(images, "../input/sign-language-for-alphabets/Sign Language for Alphabets/a", labels, 0)
temp = 97
for i in range(0, 26):
    get_images(images, os.path.join("../input/sign-language-for-alphabets/Sign Language for Alphabets/", chr(temp)), labels, i)
    temp += 1
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
images = images/255
```

```python
x_train, x_test, y_train, y_test = train_test_split(images, labels, train_size=0.9)
```

```python
x_train.shape
```

```python
data_aug = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomRotation(factor=(0.3, 0.4)),
    layers.experimental.preprocessing.RandomZoom([0.2, 0.3]),
])
```

```python
model1 = models.Sequential([
    layers.Conv2D(filters=150, kernel_size=(3, 3), padding="same", activation="relu"),
    layers.AveragePooling2D(),
    layers.Conv2D(filters=100, kernel_size=(3, 3), padding="same", activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(filters=80, kernel_size=(3, 3), padding="same", activation="relu"),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.4),
    layers.Dense(100, activation="relu"),
    layers.Dense(79, activation="relu"),
    layers.Dropout(0.2),
    layers.Dense(26, activation="sigmoid")
])
```

```python
model1.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```python
model1.fit(x_train, y_train, epochs=100)
```

```python
model1.evaluate(x_test, y_test)
```

```python
model1.save("./m1.h5")
```
